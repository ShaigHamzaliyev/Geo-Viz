```python
start_time = time.time()
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from collections import deque

# Simulate real-time signal
num_samples = 5000
timestep = 0.1
tm = np.arange(0, num_samples * timestep, timestep)
y_clean = np.sin(tm) + np.sin(tm * np.pi / 2) + np.sin(tm * -3 * np.pi / 2)
SNR = 10
ypn = y_clean + np.random.normal(0, 10 ** (-SNR / 20), len(y_clean)) - 1

print(f"Simulating real-time stream of {num_samples} samples")

# === Model Definition ===
class LinearAdaptiveFilter(nn.Module):
    def __init__(self, window_size):
        super().__init__()
        self.linear = nn.Linear(window_size, 1, bias=False)
        self.loss_fn = nn.MSELoss()
        self.optimizer = optim.SGD(self.parameters(), lr=0.001, weight_decay=1e-4)  # L2 regularization

    def forward(self, x):
        return self.linear(x)

    def adapt(self, x, target):
        pred = self.forward(x)
        loss = self.loss_fn(pred, target)
        self.optimizer.zero_grad()
        loss.backward()

        # Gradient clipping to prevent exploding weights
        torch.nn.utils.clip_grad_norm_(self.parameters(), max_norm=1.0)

        self.optimizer.step()
        return loss.item()

# === Parameters ===
window_size = 32
model = LinearAdaptiveFilter(window_size=window_size)

# Initialize input buffer
input_buffer = deque([0.0] * window_size, maxlen=window_size)

# === Real-Time Processing ===
predictions = []
losses = []

for i in range(num_samples):
    current_input = float(ypn[i])  # Noisy input
    target_output = float(y_clean[i])  # Clean desired output

    # Update buffer with new input
    input_buffer.append(current_input)

    # Skip until buffer is full
    if i < window_size:
        predictions.append(0.0)
        continue

    # Convert buffer to input tensor and normalize
    buffer_array = np.array(input_buffer)
    buffer_array = (buffer_array - buffer_array.mean()) / (buffer_array.std() + 1e-8)  # Normalize window
    x_tensor = torch.tensor([buffer_array], dtype=torch.float32)
    y_tensor = torch.tensor([[target_output]], dtype=torch.float32)

    # Predict and adapt
    with torch.no_grad():
        prediction = model(x_tensor).item()
    loss = model.adapt(x_tensor, y_tensor)

    predictions.append(prediction)
    losses.append(loss)

    if i % 500 == 0:
        print(f"Sample {i:4d} | Recent Loss: {np.mean(losses[-100:]):.6f} | Predicted: {prediction:.4f}")

print("Real-time processing complete!")

# === Evaluation ===
predictions = np.array(predictions)
mae = np.mean(np.abs(predictions - y_clean))
mse = np.mean((predictions - y_clean) ** 2)

print(f"\nFinal MAE: {mae:.6f}")
print(f"Final MSE: {mse:.6f}")

# === Plotting ===
plt.figure(figsize=(15, 8))
slices = [(0, 500, 'Early Phase'), (1000, 1500, 'Mid Phase'), (4500, 5000, 'Late Phase')]

for idx, (start, end, title) in enumerate(slices):
    plt.subplot(2, 2, idx + 1)
    plt.plot(tm[start:end], y_clean[start:end], 'g-', label='Clean Signal')
    plt.plot(tm[start:end], ypn[start:end], 'r-', label='Noisy Input', alpha=0.5)
    plt.plot(tm[start:end], predictions[start:end], 'b--', label='Prediction')
    plt.title(title)
    plt.legend()
    plt.grid(True)

# Error plot
plt.subplot(2, 2, 4)
abs_error = np.abs(predictions - y_clean)
smoothed_error = np.convolve(abs_error, np.ones(50) / 50, mode='valid')
plt.plot(smoothed_error, label='Smoothed MAE')
plt.title("Prediction Error Over Time")
plt.xlabel("Sample")
plt.ylabel("Absolute Error")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Function took {elapsed_time:.4f} seconds.")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque

# === Signal Simulation ===
num_samples = 5000
timestep = 0.1
tm = np.arange(0, num_samples * timestep, timestep)
y_clean = np.sin(tm) + np.sin(tm * np.pi / 2) + np.sin(tm * -3 * np.pi / 2)
SNR = 10
ypn = y_clean + np.random.normal(0, 10 ** (-SNR / 20), len(y_clean)) - 1

# === RNN Model ===
class RNNAdaptiveFilter(nn.Module):
    def __init__(self, input_size=1, hidden_size=16):
        super().__init__()
        self.rnn = nn.GRU(input_size, hidden_size, batch_first=True)
        self.linear = nn.Linear(hidden_size, 1)
        self.hidden_size = hidden_size
        
    def forward(self, x):
        # x shape: (batch_size, seq_len, input_size)
        out, _ = self.rnn(x)  # out shape: (batch_size, seq_len, hidden_size)
        return self.linear(out[:, -1, :])  # Last timestep

# === Parameters ===
window_size = 32  # Context window
model = RNNAdaptiveFilter(hidden_size=16)
optimizer = optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.MSELoss()

# === Real-Time Processing ===
input_buffer = deque(maxlen=window_size)
predictions = []
losses = []

for i in range(num_samples):
    current_input = ypn[i]
    target = y_clean[i]
    
    input_buffer.append(current_input)
    if i < window_size:
        predictions.append(0.0)
        continue
    
    # Prepare input: shape = (1, window_size, 1)
    x = torch.tensor(np.array(input_buffer), dtype=torch.float32).view(1, -1, 1)
    target_tensor = torch.tensor([[target]], dtype=torch.float32)
    
    # Predict and adapt
    optimizer.zero_grad()
    pred = model(x)
    loss = loss_fn(pred, target_tensor)
    loss.backward()
    optimizer.step()
    
    predictions.append(pred.item())
    losses.append(loss.item())

    if i % 500 == 0:
        print(f"Sample {i:4d} | Loss: {np.mean(losses[-100:]):.6f}")

# Evaluation
predictions = np.array(predictions)
mae = np.mean(np.abs(predictions - y_clean))
print(f"RNN MAE: {mae:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# === Signal Generation ===
num_samples = 1000
timestep = 0.1
tm = np.arange(0, num_samples * timestep, timestep)
y_clean = np.sin(tm) + np.sin(tm * np.pi / 2) + np.sin(tm * -3 * np.pi / 2)
SNR = 10
ypn = y_clean + np.random.normal(0, 10 ** (-SNR / 20), len(y_clean)) - 1

# === RLS Class ===
class RLS:
    def __init__(self, num_vars, lam=0.99, delta=1000):
        self.num_vars = num_vars
        self.A = delta * np.eye(num_vars)
        self.w = np.zeros((num_vars, 1))
        self.lam_inv = 1.0 / lam

    def add_obs(self, x, t):
        x = x.reshape(-1, 1)
        z = self.lam_inv * self.A @ x
        alpha = 1.0 / (1.0 + x.T @ z)
        error = t - float(self.w.T @ x)
        self.w += alpha * error * z
        self.A = self.lam_inv * (self.A - alpha * z @ z.T)
        return float(self.w.T @ x), error

# === Real-time Filtering Setup ===
M = 4  # number of taps (input window size)
rls = RLS(num_vars=M)

predicted = []
errors = []

# Real-time loop
for i in range(M, len(ypn)):
    x_tap = ypn[i-M:i]  # past M values
    x_tap = x_tap[::-1]  # reverse for correct time ordering
    pred, err = rls.add_obs(x_tap, ypn[i])
    predicted.append(pred)
    errors.append(err)

# Adjust arrays to align in time
t_valid = tm[M:]
y_clean_valid = y_clean[M:]
ypn_valid = ypn[M:]

# === Visualization ===
plt.figure(figsize=(12, 6))

plt.plot(t_valid, y_clean_valid, label='Clean Signal', linewidth=1)
plt.plot(t_valid, ypn_valid, label='Noisy Input (ypn)', linewidth=1, alpha=0.6)
plt.plot(t_valid, predicted, label='RLS Output (Denoised)', linewidth=1)

plt.xlabel("Time (s)")
plt.ylabel("Signal")
plt.title("RLS Real-Time Signal Denoising")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

class RLS:
    def __init__(self, num_vars, lam=0.99, delta=1.0):
        self.lam = lam
        self.num_vars = num_vars
        self.P = (1.0 / delta) * np.identity(num_vars)  # Inverse covariance matrix
        self.w = np.zeros((num_vars, 1))                # Weight vector

    def add_obs(self, x, t):
        x = x.reshape(-1, 1)  # Ensure column vector
        Px = self.P @ x
        gain = Px / (self.lam + x.T @ Px)
        error = t - float(self.w.T @ x)
        self.w += gain * error
        self.P = (self.P - gain @ x.T @ self.P) / self.lam

    def predict(self, x):
        x = x.reshape(-1, 1)
        return float(self.w.T @ x)
```
